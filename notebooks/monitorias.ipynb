### Base GPA1

````python
df = pd.read_stata("data/raw/GPA1.DTA")
df.head()
````

#### Rodando da forma usual

````python
# colGPA = b0 + b1 * hsGPA + b2 * ACT

X = df[['hsGPA', 'ACT']]
X = sm.add_constant(X)

Y = df['colGPA']

model = sm.OLS(Y, X)
results = model.fit()

print(results.summary())
````

- A interpretação do intercepto é estranha porque na amostra não existe nenhuma observação próxima de zero (não há ninguém que tenha tirado zero na escola e entrado na faculdade).
- A sigla Df Residuals do output do ```statsmodels``` significa Degrees of Freedom Residuals (Graus de Liberdade)
- A interpretação do $R^2$ diz que 17% da variação da nota da faculdade é explicada pelas notas no vestibular e do colégio.

#### Cálculando na Mão

````python
# Cálculando na mão
# colGPA = b0 + b1 * hsGPA + b2 * ACT

n = df.shape[0]
k = 3
X = sm.add_constant(df[['hsGPA', 'ACT']].to_numpy())
Y = df['colGPA'].to_numpy()

B = np.linalg.inv(X.T @ X) @ X.T @ Y
U = Y - X @ B
SIGMA_2 = (U.T @ U) / (n - k -1) # 0.1166
SIGMA = np.sqrt(SIGMA_2) # 0.3415
VAR_B = SIGMA_2 * np.linalg.inv(X.T @ X)

print(f"Os coeficientes são: {B}")
print(f"A variância dos coeficientes é: {np.diag(VAR_B)}")
print(f"O erro padrão dos coeficientes é: {np.sqrt(np.diag(VAR_B))}")
````

## Heterocedasticidade (30/06/2022)

- A presença de heterocedasticidade não afeta os nossos estimadores de ponto ($\hat{\beta}$), apenas os desvios-padrão, o que só atrapalha o teste de hipóteses.

- A primeira maneira de lidar com esse problema seria usando os desvios-padrão robustos:
  - Fórmula de White:
  
  $$\hat{Var(\hat{\beta_j})} = \frac{\sum \hat{r^2}_{ij} \times \hat{u^2}_i}{SQR^2_j} $$

  - Armazenamos os resultados da regressão em uma variável e estimar a matriz de variância-covariância corrigida usando o comando `hccm(reg)` (a versão do White é o default do R).
  - Usando `hccm(reg, type='hc0')`, temos a versão da variância do Wooldrige.
  - Usando `hccm(reg, type='hc1')`, temos a versão corrigida pelo tamanho da amostra e graus de liberdade (default do Stata).
  - Para o Python, ver este [link](https://www.statsmodels.org/dev/generated/statsmodels.regression.linear_model.RegressionResults.html)
